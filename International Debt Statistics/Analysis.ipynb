```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv("IDSData.csv")
df.fillna(0,inplace=True)

conn = sqlite3.connect(":memory:")
df.to_sql("IDSData", conn, index=False, if_exists="replace")

query = "SELECT * FROM IDSData"
result_df = pd.read_sql(query, conn)
result_df.head()
```

## We will be finding the:

1. Distinct countries

```python
query1 =  "SELECT DISTINCT `Country Name` FROM IDSData"
result_df = pd.read_sql(query1, conn)
result_df.head()
```

```python
query11 =  "SELECT COUNT(DISTINCT `Country Name`) FROM IDSData"
result_df = pd.read_sql(query11, conn)
result_df
```

2. Distinct debt indicators

```python
query2 =  "SELECT DISTINCT `Indicator Name` FROM IDSData"
result_df = pd.read_sql(query2, conn)
result_df
```

```python
query22 =  "SELECT DISTINCT `Indicator Code` FROM IDSData"
result_df = pd.read_sql(query22, conn)
result_df
```

3. Total amount of debt owed by the countries

```python
query3 = "SELECT `Country Name` ,SUM(`2024`) as Debt FROM IDSData GROUP BY `Country Name`"
result_df = pd.read_sql(query3, conn)
result_df
```

```python
query32 = "SELECT SUM(`2024`) as Debt FROM IDSData"
result_df = pd.read_sql(query32, conn)
result_df
```

4. Country with the highest debt

```python
query4 = "SELECT `Country Name`, MAX(Debt) FROM (SELECT `Country Name` ,SUM(`2024`) as Debt FROM IDSData GROUP BY `Country Name`)"
result_df = pd.read_sql(query4, conn)
result_df
```

5. Average amount of debt across indicators

```python
query5 = "SELECT `Indicator Code` ,`Indicator Name`, AVG(`2024`) as Debt_Avg FROM IDSData GROUP BY `Indicator Name` ORDER BY Debt_Avg desc"
result_df = pd.read_sql(query5, conn)
result_df
```

6. The highest amount of principal repayments

```python
Ind_Names = df.loc[df["Indicator Name"].str.contains("Principal repayments")].drop_duplicates(subset=["Indicator Code"])

Ind_Names
```

```python
query6 = "SELECT `Country Name`, `2024` as Debt FROM IDSData WHERE Debt = (SELECT MAX(`2024`) FROM IDSData WHERE `Indicator Code`='DT.AMT.DLXF.CD')"
result_df = pd.read_sql(query6,conn)
result_df
```


7. The most common debt indicator

```python
query7 = "SELECT `Indicator Name`, COUNT(*) AS Count_Entries FROM IDSData GROUP BY `Indicator Name` ORDER BY Count_Entries DESC"
result_df = pd.read_sql(query7,conn)
result_df
```
